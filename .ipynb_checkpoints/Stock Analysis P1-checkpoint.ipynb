# Business Rules

# 1️⃣ Missing Value Treatment

# (a) close_price → median per stock

```python
median_close_price = totalmerge.groupby('symbol')['close_price'].transform('median')
totalmerge['close_price'] =totalmerge['close_price'].fillna(median_close_price)
```

```python
median_close_price
```

# (b) volume → replace with 0

```python
totalmerge['volume'] = totalmerge['volume'].fillna(0)
```

```python
totalmerge
```

# (c) high_price → max(open_price, close_price)

```python
totalmerge.loc[totalmerge['high_price'].isnull(), 'high_price'] = \
    totalmerge[['open_price', 'close_price']].max(axis=1)
```

```python
totalmerge
```

# (d) low_price → min(open_price, close_price)

```python
totalmerge.loc[totalmerge['low_price'].isnull(), 'low_price'] = \
    totalmerge[['open_price', 'close_price']].min(axis=1)
```

```python
totalmerge
```

# 2.Price Sanity Checks:

# high_price ≥ open_price AND close_price

```python
totalmerge['high_price'] = totalmerge[
    ['high_price', 'open_price', 'close_price']
].max(axis=1)
```

```python
totalmerge
```

# low_price ≤ open_price AND close_price

```python
totalmerge['low_price'] = totalmerge[
    ['low_price', 'open_price', 'close_price']
].min(axis=1)
```

```python
totalmerge
```

# 3️⃣ Daily Return Calculation


```python
totalmerge['daily_return'] = (
    (totalmerge['close_price'] - totalmerge['open_price'])
    / totalmerge['open_price']
) * 100
```

```python
totalmerge
```

# 4️⃣ Trend Classification

```python
totalmerge['trend'] = totalmerge['daily_return'].apply(
    lambda x: "UP" if x > 0 else ("DOWN" if x < 0 else "NO_CHANGE")
)
```

# 1. Plot daily closing price trends per stock

Stock price is time series data

We want to see trend over time

Line plot clearly shows:

Upward trend 📈

Downward trend 📉

Sudden jumps

Crashes

```python
query = """
SELECT symbol, date, close_price
FROM stock_prices
ORDER BY symbol, date
"""

df = pd.read_sql(query, con=engine)
df['date'] = pd.to_datetime(df['date'])
```

```python
plt.figure()

colors = {
    'AAPL': 'orange',
    'GOOG': 'yellow',
    'MSFT': 'green'
}

for symbol in df['symbol'].unique():
    stock = df[df['symbol'] == symbol]
    plt.plot(stock['date'], stock['close_price'],
             label=symbol,
             color=colors.get(symbol, 'black'))

plt.title("Daily Closing Price Trends")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.show()
```

# 2️⃣ Identify Highest Volatility Stock

(Volatility = Standard Deviation of daily_return)

We are comparing multiple categories (stocks)

Each stock has one value → volatility

Bar chart is best for:

Category comparison

Ranking

Easy visual difference

```python
query = """
SELECT symbol, STDDEV(daily_return) AS volatility
FROM stock_prices
GROUP BY symbol
ORDER BY volatility DESC
"""

vol_df = pd.read_sql(query, con=engine)
```

```python
plt.figure()
plt.bar(vol_df['symbol'], vol_df['volatility'])

plt.title("Stock Volatility Comparison")
plt.xlabel("Stock Symbol")
plt.ylabel("Volatility (Std Dev)")
plt.show()
```

# 4️⃣ Moving Averages (7-Day & 30-Day)

Multiple Line Plot

Because:

We compare actual price vs smoothed price

Moving average is also time series

Helps identify:

Trend direction

Golden cross

Death cross

Noise reduction

```python
query = """
SELECT symbol, date, close_price
FROM stock_prices
ORDER BY symbol, date
"""

ma_df = pd.read_sql(query, con=engine)
ma_df['date'] = pd.to_datetime(ma_df['date'])
```

```python
ma_df['MA_7'] = ma_df.groupby('symbol')['close_price'] \
                      .rolling(7).mean() \
                      .reset_index(0, drop=True)

ma_df['MA_30'] = ma_df.groupby('symbol')['close_price'] \
                       .rolling(30).mean() \
                       .reset_index(0, drop=True)
```

```python
for symbol in ma_df['symbol'].unique():
    stock = ma_df[ma_df['symbol'] == symbol]

    plt.figure()
    plt.plot(stock['date'], stock['close_price'], label="Close Price")
    plt.plot(stock['date'], stock['MA_7'], label="7-Day MA")
    plt.plot(stock['date'], stock['MA_30'], label="30-Day MA")

    plt.title(f"{symbol} Moving Averages")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.legend()
    plt.show()
```

# 2. Days With Abnormal Volume Spikes

 Define spike = Volume > (Mean + 2 * Std Dev)

Reason: Scatter → Show specific spike days

```python
query = """
SELECT symbol, date, volume
FROM stock_prices
WHERE volume >
(
    SELECT AVG(volume) + 2 * STDDEV(volume)
    FROM stock_prices
)
"""

spike_df = pd.read_sql(query, con=engine)
spike_df['date'] = pd.to_datetime(spike_df['date'])

plt.figure()

for symbol in spike_df['symbol'].unique():
    stock = spike_df[spike_df['symbol'] == symbol]
    plt.scatter(stock['date'], stock['volume'], label=symbol)

plt.title("Abnormal Volume Spikes")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.legend()
plt.show()
```

```python
print(spike_df.head())
print(len(spike_df))
```

```python
query = """
SELECT symbol, date, volume
FROM stock_prices
WHERE volume >
(
    SELECT AVG(volume) + 1.5 * STDDEV(volume)
    FROM stock_prices
)
"""

spike_df = pd.read_sql(query, con=engine)
spike_df['date'] = pd.to_datetime(spike_df['date'])

plt.figure()

for symbol in spike_df['symbol'].unique():
    stock = spike_df[spike_df['symbol'] == symbol]
    plt.scatter(stock['date'], stock['volume'], label=symbol)

plt.title("Abnormal Volume Spikes")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.legend()
plt.show()

```

# 3️⃣ UP vs DOWN Trend Ratio Per Company

Grouped bar → Compare UP vs DOWN clearly

```python
query = """
SELECT symbol,
SUM(CASE WHEN trend = 'UP' THEN 1 ELSE 0 END) AS up_days,
SUM(CASE WHEN trend = 'DOWN' THEN 1 ELSE 0 END) AS down_days
FROM stock_prices
GROUP BY symbol
"""

trend_df = pd.read_sql(query, con=engine)

x = range(len(trend_df))

plt.figure()
plt.bar(x, trend_df['up_days'], width=0.4, label='UP')
plt.bar([i + 0.4 for i in x], trend_df['down_days'], width=0.4, label='DOWN')

plt.xticks([i + 0.2 for i in x], trend_df['symbol'])
plt.title("UP vs DOWN Trend Comparison")
plt.xlabel("Stock Symbol")
plt.ylabel("Number of Days")
plt.legend()
plt.show()
```
